```python
import pandas as pd
import numpy as np
from scipy.stats import trim_mean
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("~/Desktop/Datasets/2020_rws.csv", encoding='latin-1')
```

```python
pd.set_option('display.max_columns', None)
```

```python
data.head(20)
```

```python
m1 = trim_mean(data["On a day when you attend your employer's workplace, how many hours would you spend doing the following activities? - Preparing for work and commuting"], 0.1)
m2 = trim_mean(data["On a day when you attend your employer's workplace, how many hours would you spend doing the following activities? - Working"], 0.1)
m3 = trim_mean(data["On a day when you attend your employer's workplace, how many hours would you spend doing the following activities? - Personal and family time"], 0.1)
m4 = trim_mean(data["On a day when you attend your employer's workplace, how many hours would you spend doing the following activities? - Caring and domestic responsibilities"], 0.1)
```

```python
m5 = trim_mean(data["On a day when you do remote work, how many hours would you spend doing the following activities? - Preparing for work and commuting"], 0.1)
m6 = trim_mean(data["On a day when you do remote work, how many hours would you spend doing the following activities? - Working"], 0.1)
m7 = trim_mean(data["On a day when you do remote work, how many hours would you spend doing the following activities? - Personal and family time"], 0.1)
m8 = trim_mean(data["On a day when you do remote work, how many hours would you spend doing the following activities? - Caring and domestic responsibilities"], 0.1)
```

```python
m1, m2, m3, m4
```

```python
m5, m6, m7, m8
```

```python
categories = ["Preparing for work and commuting", "Working", "Personal and family time", "Caring and domestic responsibilities"]
trimmed_means = [m1, m2, m3, m4]

plt.bar(categories, trimmed_means, edgecolor='black')
plt.xlabel('Activities')
plt.ylabel('Mean Values(Trimmed)')
plt.title('Bar Chart of Time-Distribution for a Day of On-site Work')
plt.xticks(rotation=90)  
plt.show()
```

```python
categories = ["Preparing for work and commuting", "Working", "Personal and family time", "Caring and domestic responsibilities"]
trimmed_means = [m5, m6, m7, m8]

plt.bar(categories, trimmed_means, edgecolor='black')
plt.xlabel('Activities')
plt.ylabel('Mean Values(Trimmed)')
plt.title('Bar Chart of Time-Distribution for a Day of Remote Work')
plt.xticks(rotation=90)  
plt.show()
```

```python
trimmed_means = [m1, m2, m3, m4]

activities = [
    "Preparing and commuting",
    "Working",
    "Personal and family time",
    "Caring and domestic responsibilities"
]

fig, ax = plt.subplots()
ax.pie(trimmed_means, labels=activities, autopct='%1.1f%%')
ax.set_title('Pie Chart of Time-Distribution for a Day of On-site Work')
plt.show()
```
